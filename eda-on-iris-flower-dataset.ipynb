# Logistic Regression
## Theory of Logistic Regression

>Logistic Regression is a statistical model used for binary classification tasks. It predicts the probability that a given input belongs to a particular class by applying the logistic (sigmoid) function to a linear combination of the input features. The output is a value between 0 and 1, which represents the probability of the positive class. If the probability exceeds a threshold (usually 0.5), the model predicts class 1, otherwise class 0.

>Similar to linear regression, logistic regression is also used to estimate the relationship between a dependent variable and one or more independent variables, but it is used to make a prediction about a categorical variable versus a continuous one.

```python
# Features and target

x = df.drop(['species' ], axis = 1)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split

# Splitting the dataset
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.3, random_state = 0)

# Check shapes
print(f"x_train shape: {x_train.shape}")
print(f"x_test shape: {x_test.shape}")
```

# Model Creation

```python
from sklearn.linear_model import LogisticRegression

# Initialize the Logistic Regression model
lr = LogisticRegression()
lr.score
```

```python
lr.fit(x_train, y_train)
lr.score(x_train, y_train)
```

```python
predictions = lr.predict(x_test)
predictions 
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score (y_test, predictions)
```

The accuracy_score function from Scikit-learn calculates the proportion of correct predictions made by the model. Heres a breakdown:

y_test: The true labels (ground truth) of the test dataset.
predictions: The predicted labels generated by your model.
The accuracy score formula is:

**Accuracy** is calculated as:

Accuracy= 
TotalPredictions/
CorrectPredictions

 =0.9778 or 97.78%
 
This means that the model correctly predicted **97.78%** of the labels in the test set.
